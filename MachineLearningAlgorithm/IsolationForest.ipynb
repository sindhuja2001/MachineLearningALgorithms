```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python
data= pd.read_csv('/dataset.csv')
data.head(10)
```

```python
model=IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.2))

model.fit(data[['marks']])

print(model.get_params())
```

```python
data['scores']= model.decision_function(data[['marks']])
data['anomaly_scores']= model.predict(data[['marks']])
data[data['anomaly_scores']== -1].head()
```
